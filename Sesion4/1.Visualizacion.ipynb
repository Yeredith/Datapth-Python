```python
import matplotlib.pyplot as plt
```

```python
#Datos
x = [1,2,3,4,5]
y = [3,4,6,7,8]
plt.plot(x,y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico')
plt.show()
```

```python
#Grafico de Dispersión
plt.scatter(x,y)
plt.title('Dispersión')
plt.show()
```

```python
#Grafico de histograma
import numpy as np
datos = np.random.randn(1000)

plt.hist(datos, color = 'red', edgecolor = 'black', bins = 30)
plt.title('Histograma')
plt.show()
```

```python
categorias = ['A','B','C','D']
valores = [24,56,35,48]
plt.bar(categorias,valores, color = 'violet')
plt.xlabel('categorias')
plt.ylabel('valores')
plt.title('Grafico de barras')
plt.show()
```
